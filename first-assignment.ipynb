Function to separate even and odd numbers

```python
def separate_even_odd(lst):
    even_numbers = [num for num in lst if num % 2 == 0]
    odd_numbers = [num for num in lst if num % 2 != 0]
    return even_numbers, odd_numbers

lst = [1, 2, 3, 4, 5, 34, 5, 656, 7]
print(separate_even_odd(lst))
```

Enumerate students and categorize them by faculty

```python
students = ["Ali", "Veli", "Ayse", "Talat", "Zeynep", "Ece"]

for i, student in enumerate(students):
    faculty = "Mühendislik Fakültesi" if i < 3 else "Tip Fakültesi"
    print(f"{faculty} {i + 1}. öğrenci: {student}")
```

Set Operations

```python
set1 = {"data", "python"}
set2 = {"data", "function", "qcut", "lambda", "python", "miuul"}

if set1.issuperset(set2):
    print(set1.intersection(set2))
else:
    print(set2.difference(set1))
```

Data manipulation with seaborn and pandas

```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset("car_crashes")
numeric_columns = [f"NUM_{col.upper()}" for col in df.columns if pd.api.types.is_numeric_dtype(df[col])]
print(numeric_columns)

numeric_columns_without_no = [f"{col.replace('NUM_', '')}_FLAG" for col in numeric_columns if "NO" in col]
print(numeric_columns_without_no)

exclude_columns = ["abbrev", "no_previous"]
new_df = df[[col for col in df.columns if col not in exclude_columns]]
print(new_df)
```
